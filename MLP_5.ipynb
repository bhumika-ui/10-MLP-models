```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
data=pd.read_csv('/content/business-operations-survey-2022-information-and-communications-technology.csv',encoding='unicode_escape')
```

```python
data
```

```python
data.columns
```

```python
x=data[['description','industry','level','size','line_code']]
```

```python
y=data[['value']]
```

```python
y.size
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler , LabelEncoder
le=LabelEncoder()
x['description']=le.fit_transform(x['description'])
x['industry']=le.fit_transform(x['industry'])
x['level']=le.fit_transform(x['level'])
x['size']=le.fit_transform(x['size'])
x['line_code']=le.fit_transform(x['line_code'])
y['value']=le.fit_transform(y['value'])
x
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
from keras.utils import to_categorical
y_encoded = to_categorical(y)
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_encoded,test_size=0.2)
x_train.shape
```

```python
model=Sequential()
model.add(Dense(128,input_shape=(x_train.shape[1],),activation='relu'))
model.add(Dense(1130,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=20)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
loss,accuracy=model.evaluate(x_test,y_test)
```

```python
print('Loss',loss)
print('Accuracy',accuracy)
```
