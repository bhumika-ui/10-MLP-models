```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
data=pd.read_csv('/content/annual-enterprise-survey-2023-financial-year-provisional-size-bands.csv')
```

```python
data.columns
```

```python
x=data[['year', 'industry_code_ANZSIC', 'industry_name_ANZSIC', 'rme_size_grp',
       'variable', 'value']]
```

```python
y=data[['unit']]
```

```python
y.size
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
le=LabelEncoder()
x['industry_code_ANZSIC']=le.fit_transform(x['industry_code_ANZSIC'])
x['industry_name_ANZSIC']=le.fit_transform(x['industry_name_ANZSIC'])
x['rme_size_grp']=le.fit_transform(x['rme_size_grp'])
x['variable']=le.fit_transform(x['variable'])
x['value']=le.fit_transform(x['value'])
y['unit']=le.fit_transform(y['unit'])
x
y
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
#x_scaled=pd.DataFrame(x)
x_scaled
```

```python
from keras.utils import to_categorical
y_encoded=to_categorical(y)
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_encoded,test_size=0.2)
x_train.shape
```

```python
model=Sequential()
model.add(Dense(32,input_shape=(x_train.shape[1],),activation='relu'))
model.add(Dense(2,activation='relu'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.compile(optimizer='adagrad',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=10)
```

```python
model.compile(optimizer='adagrad',loss='categorical_crossentropy',metrics=['accuracy'])
loss,accuracy=model.evaluate(x_test,y_test)
```

```python
print('Loss',loss)
print('Accuracy',accuracy)
```
